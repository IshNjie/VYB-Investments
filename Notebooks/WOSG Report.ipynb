# Watches of Switzerland (WOSG)

Operates as a retailer of luxury watches and jewelry in the United Kingdom, Europe, and the United States.

As a retailer and distributor, they do not own the watch brands or manufacture the watches. They are referred to as an authorised dealer, officially approved by watch brands to sell its watches. This provides the customer with confidence that the products are authentic and supplied straight from the brand.

As part of the FTSE 250, WOSG have ~3000 employees across their various channels:
* Watches of Switzerland Flagship stores
* Mappin & Webb
* Goldsmiths
* Mayors
* Betteridge
* Deutsch & Deutsch

## Competitors

Watches of Switzerland will compete on multiple levels

### Luxury Watch Retailers

Other retailers will pose direct competition to WOSG. Bucherer, being a notable one, was acquired by Rolex in 2023. The idea here is to control the delivery of the watches to the consumer. Bucherer and Rolex have previously had a long standing relationship prior to the acquisition, but this just solidifies Rolex’ bid to get closer to their customer. 

### Watch Brands

Though WOSG do not own the watches or manufacture the watches, watch brands could be moving towards opening their own boutiques, and placing exclusive watches within those boutiques. This obviously disadvantages WOSG, as customers will have more choice on their high street to purchase their new [watch.](https://wristbuddys.com/en-gb/blogs/blog/new-strategy-for-rolex-exclusive-boutiques-could-become-a-reality)

### Others

Second hand markets as well as broader luxury retailers are key competitors for the brand. Chrono24 ensure the availability of watches and luxury jewellers also have the ability to sell watches as well as other items.
